```
# no 2
import matplotlib.pyplot as plt
import pandas as pd

data = pd.read_csv('Dataa.csv')

data.loc[(data[' grade'] >= 70),'Description'] = 'PASS' 
data.loc[(data[' grade'] < 70),'Description'] = 'FAIL'

data
# Tampilkan semua kolom dengan tambahan description Kolom description akan bernilai "PASS" jika grade >= 70 dan "FAIL" jika grade < 70
```

```
mean_per_student = data.groupby(["student number", " name"])[" grade"].mean()

print(mean_per_student)
print(f"rata-rata dari semua murid {data[' grade'].mean()}")

# Tampilkan kolom student number,name, dan mean (rata- rata) dari semua nilai perstudent
```
